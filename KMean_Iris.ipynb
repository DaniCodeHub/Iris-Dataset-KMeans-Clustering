# Import Initial Dataset, Modules, and Libraries #

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

data = load_iris(return_X_y=False, as_frame=False)
X = data.data
y = data.target
target_names = data.target_names
feature_names = data.feature_names
data
```

## Create Pandas DataFrame where data corresponds to the 'feature_names' ##

```python
df = pd.DataFrame(data=data['data'], columns=data['feature_names'])
df['species'] = target_names[y]
df
```

```python
KMean_model = KMeans(n_clusters=3, random_state=1000)
KMean_model
```

```python
KMean_model.fit(X)
```

```python
predictions = KMean_model.predict(X)
predictions
```

```python
## Same output as predictions above ##

labels = KMean_model.labels_
labels
```

## Import metrics from sklearn to evaluate performance ## 

```python
from sklearn.metrics import silhouette_score 

silhouette_avg = silhouette_score(X, labels)
print(silhouette_avg)

# non-negative integer is good, however, scale is -1:1, so closer to 1 represents better defined clusters #
```

```python
KMean_model.inertia_

# High inertia values indicate less compact clustering/clusters... so is 33 bad? #
```

### Lets Use Elbow method to determine the optimal number of peaks ###

```python
wcss = []
for K in range(1,11):  
    model = KMeans(n_clusters=K, random_state=1000)
    model.fit(X)
    wcss.append(model.inertia_)

print(wcss)
```

## Plot the graph to determine the elbow and see what n_clusters is most optimal ##

```python
plt.plot(range(1,11), wcss)

plt.xticks(range(1, 11))
plt.axvline(x=3, color='red', linestyle='--', linewidth=1, label='Inflection Point: Clusters = 3')


plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')

plt.legend()

plt.show()
```

```python
# From the above plot we can determine that the optimal number of clusters is '3',as there is a significant change in the rate of decrease #
```

```python
clusters = KMean_model.cluster_centers_
clusters
```

## Plot the data points and cluster centers (centroids) ##

```python
from matplotlib.patches import Patch

plt.figure(figsize=(15, 10))

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(clusters[:, 0], clusters[:, 1], c='red', marker='x')

plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('Iris Dataset KMeans clustering')

plt.colorbar(label='Cluster Label')

legend_elements = [Patch(facecolor='#E6D369', label='setosa'),
                   Patch(facecolor='#29A6A6', label='virginica'),
                   Patch(facecolor='purple', label='versicolor'),
                   Patch(facecolor='red', label='centroid')
]

plt.legend(handles=legend_elements)

plt.show()
```
